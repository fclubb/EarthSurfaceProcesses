2. What do you think will happen to this steady-state or "base" landscape if we increased rock uplift rate? How would you expect landscape relief and channel steepness to change? Write down your answers or hypothesis BEFORE you start running any code.

_To answer this question it's worth looking at your lecture notes from Week 3 and maybe your work from mountain landscapes. Remember that river profiles are aiming to adjust to their rates of erosion to balance uplift, and that the shape of the topography through time will reflect this:_

$\frac{dz}{dt} = U - K A^m S^n$

_where $\frac{dz}{dt}$ is the change in elevation through time, $U$ is uplift, and $K A^m S^n$ is erosion simulated using the stream power law._ 

_If we increase $U$, we would therefore expect rates of erosion in the channel to increase to compensate for this. This would lead to steeper channels. We would preserve high relief as the channels incise down faster into their bedrock, but this is compensated by the high rates of uplift which are overall increasing the elevation of the landscape rapidly through time._


3. What do you think would happen to this base landscape if we instead increased rock erodibility, or $K_{sp}$? How would topography and channel steepness change in this case? Write down your answers or hypothesis BEFORE you start running any code.

_Again we can look at the equation above to answer this question. If we increase $K$, we are going to increase the rates of erosion in the landscape. However unlike question 2 above, we haven't increased uplift - so the landscape is not increasing in elevation faster in this case! As channels incise down faster through weaker rock or in a wetter climate, but rock uplift doesn't change, we would therefore expect to have decreasing elevations and relief in the landscape as the rivers slowly erode the topography, and therefore lower channel steepness overall._
